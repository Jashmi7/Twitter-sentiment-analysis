```python
import re

def get_normalize(text):
    text = re.sub(r'\&\w*;', '', text)
    text = re.sub('@[^\s]+', '', text)
    text = re.sub(r'\$\w*', '', text)
    return text

# Sample usage
j = 0
for i in final_my_tweet:
    final_my_tweet[j] = get_normalize(str(i))
    j += 1

final_my_tweet
```

```python
from textblob import TextBlob
import pandas as pd

# Assuming 'new_data' is already created and contains a column 'tweets'
new_data[['Polarity', 'Subjectivity']] = new_data['tweets'].apply(
    lambda x: pd.Series(TextBlob(x).sentiment))

# Display the dataframe with the new columns
new_data
```

```python
new_data.to_csv('filtered_data.csv')
```

```python
y = new_data['Polarity']
positive = 0
negative = 0
neutral = 0

for a in y:
    if a>0:
     positive=positive+1
    elif a<0:
     negative=negative+1
    else:
     neutral=neutral+1
numberOfTweetsAnalysed= positive+negative+neutral
print('Number of Tweets Analysed = '+str(numberOfTweetsAnalysed))
print('Positive tweets = '+str(positive))
print('Negative tweets = '+str(negative))
print('Neutral tweets = '+str(neutral))

```

```python
df1 = new_data[new_data['Polarity'] > 0]
df1['sentiment'] ='Positive'
df2 = new_data[new_data['Polarity'] < 0]
df2['sentiment'] ='Negative'
df3 = new_data[new_data['Polarity'] == 0]
df3['sentiment'] ='Neutral'

new_data = pd.concat([df1,df2,df3]).sample(frac = 1)
new_data
```
